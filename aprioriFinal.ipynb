```python
def join(l1):
    res = []
    for i in range(len(l1)):
        for k in range(len(l1)):
            if i>=k:
                continue
            if set(l1[i][0:(len(l1[i])-1)]) == set(l1[k][0:len(l1[k])-1]):
                new = l1[i]
                new = np.append(new,l1[k][len(l1[k])-1])
                #print()
                res.append(new)
        
    return res
    
```

```python
import itertools
def genItem(tempres,tempcnt):
    temp = np.asarray(join(tempres))
    #ALL ITEM GENERATED
    global number
    res = []
    cnt = []
    
    exist = 0

    for i in range(len(temp)):
        pruneitem = np.array(list(itertools.combinations(temp[i],len(temp[i])-1)))
        #pruneitem = pruneitem[len(pruneitem)-1]
        exist = 0
        for y in range(len(pruneitem)):
            for x in range(len(temp)):
                c = 0
                for j in range(len(pruneitem[y])):
                    if pruneitem[y][j] in temp[x]:
                        c+=1
                    else:
                        break

                if c==len(pruneitem[y]):
                    exist+=1
                    break
            if exist < len(pruneitem):
                continue
        
        freq = isFreq(temp[i])
        #print(freq)
        if freq<min_sup:
            continue
        
        number+=1
        print(number,":",temp[i],freq)
        res.append(temp[i])
        cnt.append(freq)
        
    return np.asarray(res),cnt
```
