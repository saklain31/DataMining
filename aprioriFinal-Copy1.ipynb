```python
_end = '-'
val=1
def make_trie(words):
    #print(words)
    root = dict()
    for i in range(len(words)):
        #print("word",words[i])
        current_dict = root
        for letter in words[i]:
            current_dict = current_dict.setdefault(letter, {})
        
        #print("ccurrd",current_dict)
        
        if bool(current_dict):
            current_dict[_end] += 1
        else:
            current_dict[_end] = val
        
        #print("curr",current_dict[_end])
    return root


def in_trie(trie, word):
    current_dict = trie
    for letter in word:
        if letter in current_dict:
            current_dict = current_dict[letter]
        else:
            return 0
    else:
        if _end in current_dict:
            return current_dict[_end]
        else:
            return 0
        
        
global gtrie
gtrie = make_trie(dataset)
```

```python
def join(l1):
    res = []
    for i in range(len(l1)):
        for k in range(len(l1)):
            if i>=k:
                continue
            if set(l1[i][0:(len(l1[i])-1)]) == set(l1[k][0:len(l1[k])-1]):
                new = l1[i]
                new = np.append(new,l1[k][len(l1[k])-1])
                #print()
                res.append(new)
        
    return res
    
```

```python
import itertools
def genItem(tempres,tempcnt):
    temp = np.asarray(join(tempres))
    print("Joining",len(temp))
    
    trie = make_trie(temp)
    #ALL ITEM GENERATED
    global number
    res = []
    cnt = []
    
    res_ =[]
    
    exist = 0

    for i in range(len(temp)):
        pruneitem = np.array(list(itertools.combinations(temp[i],len(temp[i])-1)))
        #pruneitem = pruneitem[len(pruneitem)-1]
        exist = 0
        
        for y in range(len(pruneitem)):
            if in_trie(trie,pruneitem[y])>0:
                exist+=1
            if exist < len(pruneitem):
                continue                
            res_.append(temp[i])
            
        freq = isFreq(temp[i])
        #print(freq)
        if freq<min_sup:
            continue
        
        number+=1
        #print(number,":",temp[i],freq)
        res.append(temp[i])
        cnt.append(freq)
        
    print("Pruning",len(res_))
    print("FP",len(res))
    return np.asarray(res),cnt
```
